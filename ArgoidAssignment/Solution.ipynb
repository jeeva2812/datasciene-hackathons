```python
import pandas as pd
import numpy as np

data = pd.read_csv('final.csv')
print(data.shape)

data.sort_values(by = ['TransactionId'], inplace=True)
data.head()
```

```python
basket = data[:50000].groupby(['TransactionId','ItemDescription'])['NumberOfItemsPurchased'].sum().unstack().reset_index().fillna(0).set_index('TransactionId')

basket
```

```python
def encode_units(x):
    if x <= 0:
        return 0
    if x >= 1:
        return 1
basket_sets = basket.applymap(encode_units)
basket_sets
```

```python
basket_sets = basket_sets[basket_sets.sum(axis = 1) > 1]

basket_sets.shape
```

```python
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules

frq_items = apriori(basket_sets, min_support = 0.03, use_colnames = True) 
print(frq_items)
  
# Collecting the inferred rules in a dataframe 
rules = association_rules(frq_items, metric ="lift", min_threshold = 1) 
rules = rules.sort_values(['confidence', 'lift'], ascending =[False, False]) 

print(rules.shape)
rules.head()
```

```python
rules.to_csv('Assignment2.csv', index = False)
```
